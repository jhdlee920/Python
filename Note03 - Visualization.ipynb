# Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from jupyterthemes import jtplot
jtplot.style(theme='monokai')
sns.set_style('whitegrid')
```

## Example dataset

```python
# Series
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

---

## Functional Method

```python
plt.plot(x,y)
```

## Labels

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

## Creating Multi plots on the same canvas

```python
plt.plot(x,x,'r--', x, x**2,'bs',x,x**3,'g^')
```

```python
plt.subplot(121) # 1 row, 2 columns, number 1
plt.plot(x,y,'r')
plt.subplot(122) # 1 row, 2 columns, number 2
plt.plot(y,x,'b')
```

## Object Oriented Method
https://matplotlib.org/2.0.1/api/figure_api.html

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8]) # left, bottom, width, height
axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
# axes2 the set of axes starts 20% from the left, 50% from the bottom, 40% of the canvas size width,
# 30% of the canvas size height.
# , all these values here must be between 0 and 1
```

```python
fig, axes = plt.subplots(nrows=3, ncols=3)

plt.tight_layout()
```

# Seaborn - Cheatsheet

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from jupyterthemes import jtplot
jtplot.style(theme='monokai')
sns.set_style('whitegrid')
```

## Example Data

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
titanic=sns.load_dataset('titanic')
titanic.head()
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
data=pd.DataFrame({'x':np.arange(1,101),
                  'y':np.random.normal(0,4,100)})
uniform_data=np.random.rand(10,12)
```

---

## 1. Axis Grids

### 1.1. FacetGrid
* Subplot grid for plotting conditional relationships

```python
g = sns.FacetGrid(titanic, col='survived',row='sex')
g = g.map(plt.hist, "age")
```

### 1.2. factorplot
* Draw a categorical plot onto a Facetgrid

```python
sns.factorplot(x='pclass',y='survived',data=titanic)
```

### 1.3. lmplot
* Plot data and regression model fits across a FacetGrid

```python
sns.lmplot(x='sepal_width',y='sepal_length',hue='species',data=iris)
```

### 1.4.a PairGrid
* Subplot grid for plotting pairwise relationships

```python
h = sns.PairGrid(iris)
h = h.map(plt.scatter)
```

### 1.4.b pairplot
* Plot pairwise bivariate distributions

```python
sns.pairplot(iris)
```

### 1.5.a JointGrid
* Grid for bivariate plot with marginal univariate plots

```python
i = sns.JointGrid(x='x',y='y',data=data)
i = i.plot(sns.regplot, sns.distplot)
```

### 1.5.b jointplot
* Plot bivariate distribution

```python
sns.jointplot("sepal_length","sepal_width",data=iris,kind='kde')
```

## distribution plot
Show the distribution of univariate set of observations

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'], kde=False)
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=30)
```

---
## jointplot
Match up to distplots for bivariate data

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

### - kind<br>
Which allows you to affect what's actually going on inside of this joint plot. The default is a scatter plot. 

#### hex

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

#### reg
This will look a lot loke a scatterplot except Seaborn is actually goint to draw a regression line.

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='reg')
```

#### kde

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='kde')
```

---
## pairplot
To plot pairwise relationships across an entire data frame at least for the numerical columns.<br>
Also supports **a color hue** argument for categorical columns.

```python
sns.pairplot(tips)
```

### hue
for categorical data columns

```python
sns.pairplot(tips, hue='sex')
```

### palette
color this with some specific color palette

```python
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

---
## rugplot
Draws a dash mark for every points on this uniform or unique variant distribution.

```python
sns.rugplot(tips['total_bill'])
```
